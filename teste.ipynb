```python
import streamlit as st
import pandas as pd
import numpy as np

d_aluno = pd.read_csv("data/d_aluno.csv")
d_curso = pd.read_csv("data/d_curso.csv")
d_disciplina = pd.read_csv("data/d_disciplina.csv")
d_periodo = pd.read_csv("data/d_periodo.csv")
f_desempenho_academico = pd.read_csv("data/f_desempenho_academico.csv")
```

```python
d_periodo["periodo"] = (
    d_periodo["ano"].astype(str) + "/" + d_periodo["semestre"].astype(str)
)
```

```python
f_desempenho_academico.info()
```

```python
# Convertendo a coluna 'sk_d_periodo' para o tipo str em ambas as DataFrames
f_desempenho_academico['sk_d_periodo'] = f_desempenho_academico['sk_d_periodo'].astype(str)
d_periodo['sk_d_periodo'] = d_periodo['sk_d_periodo'].astype(str)
d_periodo.info()
```

```python
fato_periodo = f_desempenho_academico.merge(d_periodo, on=['sk_d_periodo'], how='inner')
```

```python
media_cra = (
    fato_periodo.groupby(["ano", "semestre", "periodo"])["cr_acumulado"]
    .mean()
    .reset_index(name="cra")
)
semestres = media_cra["semestre"].unique().tolist()
```
